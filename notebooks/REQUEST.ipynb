```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import requests
import json
import matplotlib.pyplot as plt
import time
```

```python
def get_individual_player_data(player_id):
    """ Retrieve the player-specific detailed data

    Args:
        player_id (int): ID of the player whose data is to be retrieved
    """
    base_url = "https://fantasy.premierleague.com/api/element-summary/"
    full_url = base_url + str(player_id) + "/"
    response = ''
    while response == '':
        try:
            response = requests.get(full_url)
        except:
            time.sleep(5)
    if response.status_code != 200:
        print("Response was code " + str(response.status_code))
        return None
    data = json.loads(response.text)
    return data
```

```python
def get_data():
    """ Retrieve the fpl player data from the hard-coded url
    """
    response = requests.get("https://fantasy.premierleague.com/api/bootstrap-static/")
    if response.status_code != 200:
        print("Response was code " + str(response.status_code))
        return None
    responseStr = response.text
    data = json.loads(responseStr)
    return data
```

```python
ID = 55

sdata = get_data()

IDS = len(sdata["elements"])

for ID in range(1, IDS):
    print(ID)
    fnm = sdata["elements"][ID]["first_name"]
    lnm = sdata["elements"][ID]["second_name"]
    nm = f"{fnm}_{lnm}.csv"
    nm
    print(nm)
    
    df = pd.read_csv(f'../data/legacy/combined/{nm}')

    most_recent_kickoff = df['kickoff_time'].max()

    edata = get_individual_player_data(ID)
    if edata == None:
        continue
    edata_r = list(reversed(edata["history"]))


    for i in range(len(edata_r)):
        current_kickoff = edata_r[i]["kickoff_time"]
        if current_kickoff == most_recent_kickoff:
            break
            
    dfr = pd.DataFrame(edata_r[:i])
        
    result_df = pd.concat([df, dfr], ignore_index=True)

    result_df.to_csv(f'../data/sum/{nm}')
```
